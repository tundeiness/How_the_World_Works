## I. Decision Trees: A New Paradigm

```python
# import modules needed for this task

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline


sns.set_style('darkgrid')

# import decision tree regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
# import the simulated dataset

df = pd.read_csv('simulated_data_csv')
df.head()
```

```python
def fit_and_plot_model(dataframe, model):
    features = df.drop('y', axis=1)
    target = df.y
    
    model = model.fit(features, target)
    plt.scatter(df.x, df.y)
    plt.plot(df.x, model.predict(features), 'k--')
    
    return model
```

```python
# fit and plot an unconstrained decision tree
model = fit_and_plot_model(df, DecisionTreeRegressor())

print(model)
```

```python
# evaluating the model

#get actual y
orig_y = df.y

#get predicted y
predict_y = model.predict(df.drop('y', axis=1))

#get the total difference between actual y and predicted y values
# the sum of the absolute values of the differences 

np.sum(np.absolute(orig_y - predict_y))
```

#### Summary
A prediction error of 0 in the above is not a good sign that the model did a good job. Infact the model overfitted
because it learned the training data.
Besides the tree was unconstrained: the decision tree was allowed to make as many splits as it wanted.
It ended up learning the noise along with the signal in our dataset. 

## II. Understanding Tree Structure

```python
# visualising decision tree using graphviz
# export tree structure

from sklearn.tree import export_graphviz

# import module that will allow us to render the tree in jupyter notebook

from graphviz import Source
```

```python
from Ipython.display import Image

def plot_and_save_decision_tree(model, out_file):
    # Export tree structure and save to Source object
    graph = Source( export_graphviz(model, out_file=None) )

    # Pipe to PNG bytes
    tree_png = graph.pipe(format='png')

    # Write to PNG file
    with open(out_filename,'wb') as f:
        f.write(tree_png)

    # Display image
    return Image(tree_png)
```

```python
plot_and_save_decision_tree(model, 'tree_graph.png')
```

```python
# lets display the branch/node + leaves count of the model tree

model.tree_.node_count
print('The model has {} branches/nodes'.format(model.tree_.node_count))
```

#### How to check for over-fitting in Descision trees

Since we can count the number of nodes/branches in a tree, we can also count the number of terminal nodes/
leaves. Also each leaf holds an actual prediction.


From a structural standpoint, a decision tree model is an overfit if the number of terminal node in the model
== number of observations in the dataset. This can be confirmed by using either of these 2 propeties:
- 'children_left' or,
- 'children_right'


```python
print(model.tree_.children_left)
print(model.tree_.children_right)
```

#### check length of each array:

```python
print(len(model.tree_.children_right))
print(len(model.tree_.children_left))
```

Observe that the number of leaves/terminal node in the model is **199** and model's branch/node count is also **199**

Let's also get the total number of leaf nodes:

```python
print(np.sum(model.tree_.children_right == -1))
print(np.sum(model.tree_.children_right == -1))
```

The number of observations in this dataset is **100** and from the above result we have **100** 
This shows that the decision tree model is an overfit. 

However is there a solution? See Constrained Decision Trees.

## III. Constrained Decision Trees

There are few techniques we can use to take advantage of the flexibility and complexity of decision trees while safeguarding against overfitting. 

The first solution to overfitting decision trees is to constrain the decision trees. The reason an unconstrained decision tree becomes overfit is because it's allowed to keep branching willy nilly until it ends up with one leaf per observation.

There are different ways to constrain a decision tree:

- constrain the maximum depth allowed for the tree
- set a mininum number of samples that each leaf must have


```python
# constrain/tune the maximum depth allowed for the tree

model2 = fit_and_plot_model(df, DecisionTreeRegressor(max_depth=3))
```

compare the overfit unconstrained decision tree and, the constrained decision tree by depth. It appears to approximate the sine wave fairly well.

```python
model3 = fit_and_plot_model(df, DecisionTreeRegressor(max_depth=4))
```

```python
# constraning/tunning by setting a mininum number of samples that each leaf must have

model4 = fit_and_plot_model(df, DecisionTreeRegressor(min_samples_leaf=5))
```

```python
model4 = fit_and_plot_model(df, DecisionTreeRegressor(min_samples_leaf=4))
```

```python
model4 = fit_and_plot_model(df, DecisionTreeRegressor(min_samples_leaf=6))
```

Let's continue with the decision tree you fitted and plotted in the previous exercise.

Within each **non-leaf node**, the first value indicates the split. For example, in the tree above, the very first split occurs at $x \leq 3.332$. In other words, you would follow the left branch if your observation were less than or equal to 3.332.

Within each **leaf**, the last value indicates the predicted value based on the decision path leading up to that leaf.

#### For example, let's say you had an observation $x = 3.5$ and you wanted to predict its corresponding $y$ value, or $\hat{y} = f(x) = f(3.5)$ where $f(x)$ represents the decision tree from Exercise 4.1 Part (C). Here's how that would work:

3. Split \#3: $x \leq 4.03$? **TRUE** - Go down the branch on the LEFT.

## <span style="color:RoyalBlue">Exercise 4.3 - Input Dimensions</span>

After you train a model with Scikit-Learn, if you want to use it to predict new observations/samples, you'll need to remember that the model will accept EITHER a **Pandas DataFrame** OR a **2-Dimensional NumPy Array**. (In fact, under the hood, Scikit-Learn actually converts a Pandas DataFrame into a 2-Dimensional NumPy Array first!)

The key word there is **2-Dimensional**. The first dimension represents the "rows" or individual observations of your dataset. The second dimension represents the "columns" or input features of your dataset. Even if you only have one input feature (like we do in this simulation), the model will still expect a 2-dimensional array.

#### For example...
* If you have 1 observation and 1 input feature, you'll need a $1 \times 1$ array.
* If you have 100 observations and 1 input feature, you'll need a $100 \times 1$ array.
* If you have 1 observation and 5 input features, you'll need a $1 \times 5$ array.
* If you have 100 observation and 5 input features, you'll need a $100 \times 5$ array.
